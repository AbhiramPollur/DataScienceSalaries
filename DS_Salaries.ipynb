```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("/Users/mac/Desktop/CSV Files/Salary_Data.CSV")
```

```python
data_columns = data.columns
print(data_columns)
```

```python
column_age = data['Age']
colors = ['#7DF9FF', '#00FFEF', '#08E8DE', '#48D1CC', '#00CED1']
```

```python
column_education = data['Education Level']
for education in column_education.unique():
    print(education)
```

```python
avg_salary_education = data.groupby(data['Education Level'])['Salary'].mean()
print(avg_salary_education)
avg_salary_education.plot(kind='bar', color=colors)
```

```python
column_gender = data['Gender']
for gender in column_gender.unique():
    print(gender)
```

```python
avg_salary_gender = data.groupby(data['Gender'])['Salary'].mean()
print(avg_salary_gender)
avg_salary_gender.plot(kind='bar', color=colors) 
```

```python
avg_salary_male = avg_salary_gender['Male']
avg_salary_female = avg_salary_gender['Female']
male_female = (avg_salary_male/avg_salary_female)
print(male_female)
```

```python
column_jobtitle = data['Job Title']
for jobtitle in column_jobtitle.unique():
    print(jobtitle)
```
